```python
from cloudbusting.data import TrainImageList
import matplotlib.pyplot as plt
```

## load data from data directory

```python
trainImageList = TrainImageList('../../data')
```

```python
def plot_image(i):
    fig, ax = plt.subplots(dpi=200)
    image = trainImageList[i]
    image.plot(ax)
    plt.show()
```

## show first image in sequence

```python
plot_image(0)
```

## show image by name

```python
plot_image('159985f')
```
